Program for heart disease prediction AI/ML model

```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, BaggingClassifier, VotingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SelectKBest, f_classif
```

Data collection and Pre-procesing

```
# loading our heart.csv file to a panda dataframe
heart_data = pd.read_csv('/content/heart.csv')
heart_data.shape
```

```
# getting some info about our dataset
heart_data.info()
```

```
# statistical measure of our data
heart_data.describe()
```

```
# checking the distribution of target variable
heart_data['target'].value_counts()
```

```
# Data preprocessing
#1. Handling missing values
imputer = SimpleImputer(strategy='median')
heart_data_imputed = pd.DataFrame(imputer.fit_transform(heart_data), columns=heart_data.columns)
```

```
# Splitting features and target
X = heart_data_imputed.drop(columns='target', axis=1)
Y = heart_data_imputed['target']
print(X)
print(Y)
```

```
#Feature scaling(Normalization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X)
```

```
from sklearn.feature_selection import SelectKBest, f_classif
# automatic feature selection
print('Selecting best number of features using cross validation')
best_k = 0
best_score = 0

# Temporarily store the best selector found within the loop
final_selector_for_X_new = None

for k in range(5, X.shape[1] + 1):
   temp_selector = SelectKBest(score_func=f_classif, k=k)
   X_temp = temp_selector.fit_transform(X_scaled, Y)
   model_cv = LogisticRegression(max_iter=1000)
   scores = cross_val_score(model_cv, X_temp, Y, cv=10, scoring='f1')
   if scores.mean() > best_score:
     best_score = scores.mean()
     best_k = k
     final_selector_for_X_new = temp_selector # Store this selector if it leads to the best score

print(f'\n Best number of features (k): {best_k}')
print(f'\n Cross validated F1-score for best k: {best_score:.3f}')

# Initialize and fit the global 'selector' variable with the best_k found
# This ensures the 'selector' object used later for prediction has the correct configuration
selector = SelectKBest(score_func=f_classif, k=best_k)
X_new = selector.fit_transform(X_scaled, Y)
```

```
# splitting data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X_new, Y, test_size=0.14, stratify=Y, random_state=2)
print (X.shape, X_train.shape, X_test.shape)
print(f'\nOriginal Shape: {X_scaled.shape}, Training Shape: {X_train.shape}, Testing Shape: {X_test.shape}')
```

```
# Define models
log_reg = LogisticRegression(max_iter=1000, random_state=2)
dt_clf = DecisionTreeClassifier(max_depth=8, random_state=2)
rf_clf = RandomForestClassifier(n_estimators=200, random_state=2)
gb_clf = GradientBoostingClassifier(n_estimators=200, learning_rate = 0.05, random_state=2)
ada_clf = AdaBoostClassifier(n_estimators=200,learning_rate = 0.05, random_state=2)
# voting classifier combining major models
voting_clf = VotingClassifier(estimators=[('lr', log_reg), ('dt', dt_clf), ('rf', rf_clf), ('gb', gb_clf),], voting='soft')
models = {'Logistic Regression': log_reg, 'Decision Tree': dt_clf, 'Random Forest': rf_clf, 'Gradient Boosting': gb_clf, 'AdaBoost': ada_clf, 'Voting Classifier': voting_clf}
```

```
# Model Training
best_model = None
highest_f1 = -1
print('\n Training and evaluation results forall models')
for model_name, model in models.items():
#model_name = DecisionTreeClassifier()
 model.fit(X_train, Y_train)


```

```
# Model evaluation
X_test_prediction = model.predict(X_test)

# accuracy score
# accuracy on Training data
X_train_prediction = model.predict(X_train)
accuracy_train = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data:', accuracy_train)

#accuracy on test data
X_test_prediction = model.predict(X_test)
accuracy_test = accuracy_score(Y_test, X_test_prediction)
#print('Accuracy on Test data:', accuracy_test)

#precision
precision_train = precision_score(X_train_prediction, Y_train)
print('Precision on training data:', precision_train)
precision_test = precision_score(Y_test, X_test_prediction)
#print('Precision on testing data:', precision_test)

#recall
recall_train = recall_score(X_train_prediction, Y_train)
print('Recall score of training data:', recall_train)
recall_test = recall_score(Y_test, X_test_prediction)
#print('Recall score of testing data:', recall_test)

#f1 score
f1_train = f1_score(X_train_prediction, Y_train)
print('F1 score of training data:', f1_train)
f1_test = f1_score(Y_test, X_test_prediction)
#print('F1 score of training data:', f1_test)

print(f'\n**{model_name}**')
print(f'Accuracy on testing data: {accuracy_test:.4f}')
print(f'Precision on testing data: {precision_test:.4f}')
print(f'Recall on testing data: {recall_test:.4f}')
print(f'F1 score on testing data: {f1_test:.4f}')
print(f'\nClassification Report for {model_name}:\n {classification_report(Y_test, X_test_prediction)}')
if f1_test > highest_f1:
  highest_f1 = f1_test
  best_model = model
```

```
# Building a Predictive System
print('\n Enter the following details to predict heart disease')
feature_name = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
input_data = []
for feature in feature_name:
  value = float(input(f'Enter {feature}: '))
  input_data.append(value)
# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
input_data_scaled = scaler.transform (input_data_reshaped)

# Apply feature selection to the input data
input_data_final = selector.transform(input_data_scaled)

prediction = best_model.predict(input_data_final)
print(prediction)

if(prediction[0] == 0):
  print('Your heart is healthy')
else:
  print('You have a heart disease')
```
